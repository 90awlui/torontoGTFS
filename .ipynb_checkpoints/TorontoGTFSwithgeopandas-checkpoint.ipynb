# Unique Routes for Plotting 
It doesn't look like it's easy to plot these routes by colours. I want to make it as efficient as possible to plot our routes. It all looks pretty scattered right now. Let's do some work to make some kind of hierarchy for plotting the routes. This hierarchy doesn't necessarily apply for doing frequency etc.

1. Route short name / long name / Route ID (from routes)
2. Each route_id has multiple shape_id's (from trips)
3. Each shape_id has multiple shape_lat and shape_lon (from shapes)

As a side note, I also tried mashing everything into one giant dataframe but that didn't seem like such a good way to go
in terms of efficiency... Probaby doubling down on routes a LOT. That said we need to somehow account for different 
services

Okay thought #2, routes seems pretty useless by itself so let's mash it together with trips and use a combo of trips/routes as our base. Then we can at least do stuff according to service IDs

```python
trips_routes = pd.merge(trips,routes,how='left',on='route_id')
trips_routes.drop(['trip_short_name','route_desc','route_url','route_text_color'],axis=1,inplace=True)
trips_routes
```

```python
#i want to see if shapes can be assigned to more than 1 route
unique_shapes = trips_routes['shape_id'].unique()
len(unique_shapes)
```

```python
#sample: for shape_id 715921, what are all the unique route_id's?
trips_routes[trips_routes['shape_id']==715921]['route_id'].unique()
```

```python
#now we repeat for the sample across all shapes
check_unique = [len(trips_routes[trips_routes['shape_id']==shape]['route_id'].unique()) for shape in unique_shapes]
max(check_unique)
```

Luckily, every shape is unique, so we could actually just replace shape_id with the polyline associated with that shape.

## Generate polyline for every shape_id

```python
#we create a shape list that compresses all the shape sequences 
#into one shape 
tempshapes = shapes['shape_id'].unique()
unique_shapes = pd.DataFrame(columns=['shape_id','polyline','shape_dist'])
for shape in tempshapes:
    tempdf = shapes[shapes['shape_id']==shape]
    polyline = list(zip(tempdf['shape_pt_lat'],tempdf['shape_pt_lon']))
    dist = max(tempdf['shape_dist_traveled'])
    unique_shapes.loc[len(unique_shapes)]=[shape,polyline,dist]
    
unique_shapes.head()
```

```python
trips_routes = pd.merge(trips_routes,unique_shapes,how='left',on='shape_id')
```

## Calculating Headways

So we want to adjust the line thicknesses of based on headway of the route
should we do both directions and then assume an even split in both?


```python
#let's just look at ones where stop_sequence == 1 
stops_seq_1 = stop_times[stop_times['stop_sequence']==1]
stops_seq_1.head()
```

```python
trips_routes.head()
```

```python
#now we want to merge this with the trips _routes table because we want to be able to work on the routes
#we're also getting rid of any irrelevant columns to try and save some space.
trips_seq_1 = pd.merge(trips_routes,stops_seq_1,how='inner',on='trip_id')
trips_seq_1.dropna(axis=1,how='all',inplace=True)
trips_seq_1.drop(columns=['trip_headsign','shape_id','block_id','agency_id','polyline','stop_sequence'],inplace=True)
trips_seq_1.info()
```

```python
trips_seq_1.head()
```

```python
#now we want to separate out each service id to look at separately. 
trips_serv_id_1 = trips_seq_1[trips_seq_1['service_id']==1]
trips_serv_id_2 = trips_seq_1[trips_seq_1['service_id']==2]
trips_serv_id_3 = trips_seq_1[trips_seq_1['service_id']==3]
```

```python
trips_serv_id_3.info()
```

a pivot table isn't that usefule we should probably just use a multiindex dataframe with service id then hours. does that make sense? maybe service id for weekends doesn't really have a peak. 

```python
#we want to extract the departure hour for each trip. 
# This will cut off trips that start before the peak but still service during the peak,
# and includes trips that being at the very end of the peak, and do not service the
# peak. I think this generally comes out in the wash though and isn't a huge deal.

# Treats direction_id 0 and 1 as same. Will need to divide numbers by 2 to get headway or average service per trip

for each in [trips_serv_id_1,trips_serv_id_2,trips_serv_id_3]:
    each['depart_hour'] = [int(x[:x.find(':')]) for x in each['departure_time']]
```

```python
unique_routes = sorted(trips_routes['route_short_name'].unique())
unique_routes
```

```python
def headway(df,route,peak):
    #This function calculates the headway for a route during a peak given (dataframe,route,peak)
    num_trips = len(df[(df['route_short_name']==route) & (df['depart_hour'].isin(peak))])
    num_hours = len(df.loc[(df['route_short_name']==route) & (df['depart_hour'].isin(peak)),'depart_hour'].unique())
    if num_hours == 0:
        return 0
    else:
        return round(60 / ((num_trips / 2) / num_hours),1)
```

```python
#headways adjusted for 2 directions 
serv1_headways = pd.DataFrame(index=unique_routes,columns=['AM','MD','PM','NON'])
serv2_headways = pd.DataFrame(index=unique_routes,columns=['AM','MD','PM','NON'])
serv3_headways = pd.DataFrame(index=unique_routes,columns=['AM','MD','PM','NON'])

#for each service_id 
for serv,servdata in [[serv1_headways,trips_serv_id_1],
                     [serv2_headways,trips_serv_id_2],
                     [serv3_headways,trips_serv_id_3]]:
    #calculate the headway (minutes) for all routes 
    for col,peak in [['AM',amPeak],['MD',mdPeak],['PM',pmPeak],['NON',nonPeak]]:
        serv[col] = [headway(servdata,x,peak) for x in unique_routes]
```

# Plotting with Folium

```python
import folium
from folium import plugins
```

```python
m = folium.Map(location=[43.775923, -79.346974],tiles='Stamen Toner',zoom_start=10)
busFG = folium.FeatureGroup(name='Bus Routes')
subwayFG = folium.FeatureGroup(name='Subway Routes')
stopsFG = folium.FeatureGroup(name='Stops')
wheelchair2FG = folium.FeatureGroup(name='Wheelchair Inaccessible')
busFG.add_to(m)
subwayFG.add_to(m)
stopsFG.add_to(m)
wheelchair2FG.add_to(m)
folium.map.LayerControl().add_to(m)
```

```python
#Lesson 1 - anything exceeding like 4MB worth of visualizations is going to not show up in folium so have to save it.
#Lesson 2 - if there's ' or & in the popup it's going to screw up so we have to parse_html = True first 

#1m33s with subset = shape_id
for index, row in trips_routes.drop_duplicates(subset='shape_id').iterrows():
    popup = folium.Popup(str(row['route_short_name'])+' '+row['route_long_name'],parse_html=True)
    if int(row['route_short_name']) <= 4:
        route = folium.PolyLine(locations=row['polyline'],color='#' + row['route_color'].lower(),weight=3,popup=popup,opacity=0.8)
        route.add_to(subwayFG)
    else:
        route = folium.PolyLine(locations=row['polyline'],color='#' + row['route_color'].lower(),weight=2,popup=popup,opacity=0.8)   
        route.add_to(busFG)

#plotting wheelchair inaccessible 
for index, row in trips_routes[trips_routes['wheelchair_accessible']==2].drop_duplicates(subset='shape_id').iterrows():
    popup = folium.Popup(str(row['route_short_name'])+' '+row['route_long_name'],parse_html=True)
    route = folium.PolyLine(locations=row['polyline'],color='#' + row['route_color'].lower(),weight=2,popup=popup,opacity=0.8)   
    route.add_to(wheelchair2FG)
#subwayRoutes.add_to(m)
#busRoutes.add_to(m)

```

```python
for index, row in stops[:400].iterrows():
    stop = folium.Marker(location=(row['stop_lat'],row['stop_lon']))
    stop.add_to(stopsFG)
```
